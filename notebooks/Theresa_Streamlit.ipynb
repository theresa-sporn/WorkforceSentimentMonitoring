```python
from WorkforceSentimentMonitoring.data import get_data, merge
```

```python
submission, train, test = get_data()
```

```python
df = merge(submission, train, test)
```

```python
df = df[:100]
```

```python
df_new = df[['overall', 'review']]
```

```python
#create positive, negative, neurtal dataframe
df_new.head(10)
```

```python
df_neg = df_new[df_new['overall']<3]
df_neg = df_neg['review']
df_neg
```

```python
df_pos = df_new[df_new['overall']>3]
df_pos = df_pos['review']
df_pos
```

```python
df_neu = df_new[df_new['overall']==3]
df_neu = df_neu['review']
df_neu
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = df_neu
```

```python
unique_string_neg=(" ").join(df_neg)
wordcloud_neg = WordCloud(width = 500, height = 500).generate(unique_string_neg)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud_neg)
plt.axis("off")
plt.show()
```

```python
unique_string_pos=(" ").join(df_pos)
wordcloud_pos = WordCloud(width = 500, height = 500, max_words=10).generate(unique_string_pos)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud_pos)
plt.axis("off")
plt.show()
```

```python
unique_string_neu=(" ").join(df_neu)
wordcloud_neu = WordCloud(width = 500, height = 500).generate(unique_string_neu)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud_neu)
plt.axis("off")
plt.show()
```
