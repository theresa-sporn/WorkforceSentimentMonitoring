```python
from WorkforceSentimentMonitoring.encoders import Preprocessor

preprop = Preprocessor()
pos_preprocessed = {}
neg_preprocessed = {}
for category in categories:
    pos_prep = preprop.fit_transform(pd.DataFrame(pos[f'{category}']['review']))
    pos_preprocessed[f'{category}'] = pos_prep
    neg_prep = preprop.fit_transform(pd.DataFrame(neg[f'{category}']['review']))
    neg_preprocessed[f'{category}'] = neg_prep
```

```python
from WorkforceSentimentMonitoring.preprocessing import tokenize

pos_tokenized_dict = {}
neg_tokenized_dict = {}
for category in categories:
    pos_tokenized_text = []
    for text in pos_preprocessed[f'{category}']['review']:
        pos_tokenized_text.append(tokenize(text))
    pos_tokenized_dict[f'{category}'] = tokenized_text
    neg_tokenized_text = []
    for text in neg_preprocessed[f'{category}']['review']:
        neg_tokenized_text.append(tokenize(text))
    neg_tokenized_dict[f'{category}'] = tokenized_text
```
